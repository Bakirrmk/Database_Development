<h3>Модели Базы Данных</h3>

Существует несколько типов моделей баз данных, каждая из которых представляет способ организации данных и их взаимодействия. Вот основные из них:

1) Иерархическая модель

Иерархическая модель данных использует древовидную структуру для представления данных, где каждый элемент (запись) связан с одним родительским элементом и может иметь несколько дочерних элементов. Это напоминает структуру дерева.

Принципы:

Каждый элемент данных связан с другим элементом через родительскую и дочернюю связь.
Все данные организованы в виде дерева, что ограничивает гибкость при изменении структуры.

Пример: В системе управления документами данные могут быть организованы как дерево:
Родительский элемент (например, "Отчеты") может содержать дочерние элементы ("Финансовые отчеты", "Технические отчеты") и так далее.

2) Сетчатая модель

Сетчатая модель является развитием иерархической модели. В отличие от иерархической модели, где данные могут иметь только один родительский элемент, в сетчатой модели каждый элемент может иметь несколько родительских элементов, образуя сложную сеть взаимосвязей.
    
Принципы:

Используется структура графа, где каждый элемент может иметь несколько связей с другими элементами.
Связи могут быть многоуровневыми, что дает больше гибкости, чем в иерархической модели.

Пример: В системе управления проектами сотрудники могут быть назначены на несколько проектов, а проекты могут включать несколько сотрудников. Элементы "сотрудник" и "проект" могут быть связаны друг с другом через сетевую структуру.

3) Реляционная модель

Реляционная модель — это основа большинства современных систем управления базами данных (СУБД), таких как MySQL, PostgreSQL, Oracle, MS SQL Server и другие.

Принципы реляционной модели:

Данные хранятся в таблицах (или отношениях), где строки — это записи, а столбцы — атрибуты данных.
Каждая таблица имеет ключ (например, первичный ключ), который уникально идентифицирует каждую строку.
Связи между таблицами устанавливаются через внешние ключи. Это позволяет связывать таблицы и работать с ними как с единым целым.

Пример: Представим базу данных для интернет-магазина с таблицами:

Customers (Клиенты) — данные о клиентах.
Products (Товары) — информация о товарах.
Orders (Заказы) — сведения о заказах клиентов.
Связь: каждый заказ связан с клиентом (через внешний ключ), а также содержит ссылки на товары (через промежуточную таблицу OrderItems).

4) Объектно-ориентированная модель

В объектно-ориентированной модели данные представляют собой объекты, которые инкапсулируют как атрибуты (данные), так и методы (функции для работы с данными). Эта модель предназначена для использования с объектно-ориентированным программированием.

Принципы:

Данные представлены в виде объектов.
Объекты могут наследовать характеристики и методы от других объектов (наследование).
Объекты могут быть связаны друг с другом через ассоциации или агрегации.

Пример: В системе учета заказов можно моделировать объекты "Клиент", "Заказ", "Товар", где каждый объект будет иметь методы, такие как создание заказа, изменение статуса и т.д.

5) Документная модель (NoSQL)

Документные базы данных представляют данные в виде документов (чаще всего в формате JSON или BSON), где каждый документ может содержать набор ключ-значение с различной структурой.

Принципы:

Данные хранятся в документах (например, JSON), которые могут иметь разнообразную структуру.
Каждому документу присваивается уникальный идентификатор, и доступ к данным осуществляется через этот ID.

Пример: База данных для хранения информации о пользователях, где каждый пользователь может иметь разные поля: имя, возраст, предпочтения, история покупок и т.д. Все это хранится в одном документе.

6) Графовая модель (NoSQL)

Графовая модель данных используется для представления данных в виде графов, где данные хранятся как вершины (объекты) и ребра (связи между объектами).

Принципы:

Каждый объект данных является вершиной графа, а связи между объектами — ребрами.
Эффективна для работы с взаимосвязанными данными, такими как социальные сети, рекомендации и т.д.

Пример: Социальная сеть, где каждый пользователь является вершиной, а ребра представляют дружеские отношения между ними.

<h3>Логическая и Физическая структура БД</h3>

- Логическая структура базы данных

Логическая структура базы данных описывает организацию данных на концептуальном уровне, не затрагивая конкретные технологии хранения и реализации. Она определяет, как данные будут структурированы, какие таблицы и связи между ними будут использоваться, и каким образом данные будут представлены пользователю или приложению. Логическая структура базы данных зависит от требований бизнеса и специфики хранения информации, но не зависит от технологий, которые будут использоваться для реализации.

- Физическая структура базы данных

Физическая структура базы данных описывает, как данные будут храниться в памяти компьютера или на диске. Это уровень, который решает, как именно информация будет организована и оптимизирована для быстрого доступа и эффективного использования ресурсов (например, памяти и процессора). Физическая структура определяет, как будет реализована логическая структура на практике с точки зрения хранения, индексации, резервного копирования и других аспектов.

<h3>Схема данных</h3>

Схема данных — это абстрактное представление структуры базы данных, которое описывает, как организованы и как взаимосвязаны данные в базе. Она определяет все элементы, которые могут быть сохранены в базе данных, и их взаимное расположение. Схема данных является основой для проектирования и реализации базы данных и описывает следующие ключевые аспекты:

Сущности: Это объекты или понятия, которые имеют значение для бизнеса или области применения и которые необходимо хранить в базе данных. Пример: "Клиенты", "Продукты", "Заказы".
Атрибуты: Это характеристики сущностей. Атрибуты описывают свойства сущности. Например, для сущности "Клиент" атрибутами могут быть "Имя", "Адрес", "Телефон".

Связи (или отношения): Определяют, как сущности взаимодействуют друг с другом. Связи могут быть:
Один к одному (1:1): Одна запись в первой таблице связана с одной записью во второй.
Один ко многим (1:M): Одна запись в первой таблице может быть связана с несколькими записями во второй.
Многие ко многим (M:M): Несколько записей в первой таблице могут быть связаны с несколькими записями во второй.

Схема данных, как правило, создаётся в виде диаграммы, где сущности изображаются в виде прямоугольников, а связи между ними — линиями, соединяющими соответствующие сущности. Диаграммы, такие как ER-диаграммы (диаграммы "сущность-связь"), часто используются для визуального представления схемы данных.

<h3>Ключи в базе данных</h3>

1) Первичный ключ (Primary Key)

Первичный ключ — это уникальный идентификатор для каждой строки в таблице. Он используется для гарантии того, что каждая запись в таблице может быть однозначно идентифицирована.

Свойства первичного ключа:

Уникальность: Значение первичного ключа в каждой строке должно быть уникальным, то есть ни одна строка не может иметь одинаковое значение в этом поле.
Не может быть NULL: Первичный ключ не может содержать пустых значений (NULL).
Обычно один первичный ключ: В таблице может быть только один первичный ключ, который может состоять из одного или нескольких столбцов (если это составной ключ).
Пример:

В таблице Клиенты первичным ключом может быть ID клиента, потому что каждый клиент имеет уникальный идентификатор.
В таблице Заказы первичным ключом может быть ID заказа, чтобы каждый заказ можно было идентифицировать.

2) Вторичный ключ (Foreign Key)

Внешний ключ — это атрибут (или группа атрибутов) в одной таблице, который ссылается на первичный ключ другой таблицы. Внешний ключ используется для установления связи между двумя таблицами, обеспечивая целостность данных.

Свойства внешнего ключа:

Ссылка на первичный ключ другой таблицы: Внешний ключ указывает на первичный ключ в другой таблице, что позволяет связать записи между таблицами.
Целостность данных: Внешний ключ гарантирует, что значения в столбце внешнего ключа соответствуют значениям в столбце первичного ключа другой таблицы (или являются NULL, если разрешено).
Может содержать NULL: В отличие от первичного ключа, внешний ключ может быть NULL, что означает отсутствие связи.
Пример:

В таблице Заказы может быть внешний ключ ID клиента, который ссылается на ID клиента в таблице Клиенты. Это позволяет каждому заказу быть связанным с определённым клиентом.
В таблице Заказанные товары может быть внешний ключ ID заказа, ссылающийся на ID заказа в таблице Заказы.

3) Вторичный ключ (Secondary Key)

Вторичный ключ — это атрибут (или группа атрибутов), который используется для организации быстрого поиска и сортировки данных. Вторичный ключ может не быть уникальным и может содержать дублирующиеся значения.

Свойства вторичного ключа:

Не обязательно уникален: В отличие от первичного ключа, вторичный ключ может содержать дублирующиеся значения.
Создание индекса: Вторичный ключ может быть использован для создания индекса, который ускоряет поиск данных по определённому столбцу.
Может использоваться для поиска: Он не обязательно ссылается на первичный ключ другой таблицы, а просто обеспечивает быстрый доступ к данным.
Пример:

В таблице Клиенты может быть вторичный ключ Телефон, если нужно быстро найти клиента по его номеру телефона. Этот ключ не уникален, потому что несколько клиентов могут иметь одинаковые номера телефонов (например, семейные номера).
В таблице Продукты может быть вторичный ключ Категория, чтобы быстро найти все продукты определённой категории.

<h3>Сложные запросы</h3>

Сложные запросы в SQL включают в себя более продвинутые операции, такие как объединение таблиц, подзапросы, использование агрегатных функций, оконных функций и другие техники для извлечения и манипуляции данными в реляционных базах данных. Они позволяют работать с большими объемами данных, проводить анализ и получать более специфическую информацию. 

<b>Oператоры</b>

1) Оператор JOIN

Оператор JOIN используется для объединения данных из двух или более таблиц на основе какого-либо условия (обычно через ключевые поля). Это один из наиболее важных операторов для работы с нормализованными базами данных, где данные распределены между разными таблицами.

Типы JOIN:

INNER JOIN: Возвращает только те строки, которые имеют совпадения в обеих таблицах.
LEFT JOIN (или LEFT OUTER JOIN): Возвращает все строки из левой таблицы и соответствующие строки из правой таблицы. Если нет совпадений, в правой таблице будут NULL значения.
RIGHT JOIN (или RIGHT OUTER JOIN): Возвращает все строки из правой таблицы и соответствующие строки из левой таблицы.
FULL JOIN (или FULL OUTER JOIN): Возвращает все строки из обеих таблиц, даже если нет совпадений.

Пример:
SELECT Клиенты.Имя, Заказы.Дата, Заказы.Сумма
FROM Клиенты
INNER JOIN Заказы ON Клиенты.ID = Заказы.ID_клиента;

Этот запрос возвращает имена клиентов и их заказы, где для каждого клиента существует хотя бы один заказ.

2) Оператор HAVING

Оператор HAVING используется для фильтрации данных, которые были сгруппированы с помощью GROUP BY. В отличие от WHERE, который фильтрует данные до группировки, HAVING применяется уже после.

Пример:
SELECT ID_клиента, SUM(Сумма) AS Общая_сумма
FROM Заказы
GROUP BY ID_клиента
HAVING SUM(Сумма) > 500;

Этот запрос выводит клиентов, у которых общая сумма заказов больше 500.

3) Оператор GROUP BY

Оператор GROUP BY используется для группировки строк, которые имеют одинаковые значения в одном или нескольких столбцах. Он часто используется с агрегатными функциями, такими как COUNT, SUM, AVG, MAX, MIN, для выполнения вычислений по группам.

Пример:

SELECT ID_клиента, COUNT(*) AS Количество_заказов
FROM Заказы
GROUP BY ID_клиента;

Этот запрос возвращает количество заказов для каждого клиента.

4) Оператор ORDER BY

Оператор ORDER BY используется для сортировки результата запроса по одному или нескольким столбцам. По умолчанию сортировка выполняется в порядке возрастания (ASC), но можно указать порядок убывания (DESC).

Пример:
SELECT Имя, Дата_заказа
FROM Заказы
ORDER BY Дата_заказа DESC;

Этот запрос возвращает заказы, отсортированные по дате в убывающем порядке.

5) Оператор CAST

Оператор CAST используется для преобразования одного типа данных в другой. Это полезно, если необходимо привести данные к определенному типу для выполнения арифметических операций, сравнения или вывода.

Пример:
SELECT Имя, CAST(Сумма AS DECIMAL(10,2)) AS Сумма_в_формате
FROM Заказы;

Этот запрос преобразует столбец Сумма в тип DECIMAL с двумя знаками после запятой. Это полезно, если данные представлены в виде строк или целых чисел, а для вывода или вычислений нужно использовать тип с плавающей точкой.

6) Оператор CASE

Оператор CASE используется для реализации условной логики в SQL-запросах. Он позволяет выполнять различные действия в зависимости от значений в столбцах или выражениях. Это аналог оператора IF-ELSE в других языках программирования.

Существует два типа оператора CASE:

С простым выражением (CASE expression), когда проверяется одно выражение и соответствующие условия.
С поисковым выражением (CASE WHEN), когда можно указать несколько условий для проверки.

Простой CASE используется, когда вы хотите сравнить столбец или выражение с несколькими возможными значениями.
Пример:
SELECT Имя, 
       CASE Статус
           WHEN 'A' THEN 'Активный'
           WHEN 'I' THEN 'Неактивный'
           ELSE 'Неизвестный статус'
       END AS Статус_клиента
FROM Клиенты;

Этот запрос:
Если значение в столбце Статус равно 'A', то возвращает "Активный".
Если значение равно 'I', то возвращает "Неактивный".
В случае других значений возвращает "Неизвестный статус".

7) Операторы IS NULL и IS NOT NULL

Операторы IS NULL и IS NOT NULL в SQL используются для проверки того, является ли значение в столбце NULL (пустым) или не является им. Это важно, потому что NULL в базе данных представляет собой отсутствие значения, а не пустое или нулевое значение.

Оператор IS NULL проверяет, является ли значение в столбце NULL (пустым).

Пример:
SELECT Имя, Дата_рождения
FROM Клиенты
WHERE Дата_рождения IS NULL;

Этот запрос возвращает всех клиентов, у которых Дата_рождения не указана (то есть значение в этом столбце NULL).

S NULL — это специальная проверка, которая возвращает TRUE, если в столбце находится значение NULL.
В SQL NULL не равен никакому другому значению, даже самому себе. Поэтому для проверки на NULL нельзя использовать операторы равенства (например, = NULL), так как они всегда возвращают FALSE.

Оператор IS NOT NULL используется для проверки, что значение в столбце не NULL.

Пример:
SELECT Имя, Дата_рождения
FROM Клиенты
WHERE Дата_рождения IS NOT NULL;

Этот запрос возвращает всех клиентов, у которых Дата_рождения указана (то есть значение в этом столбце не NULL).

IS NOT NULL проверяет, что в столбце есть какое-то значение (не пустое).
Этот оператор используется, если нужно исключить строки, где значение отсутствует.
